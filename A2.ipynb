```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import altair
import os
import cartopy.crs as ccrs
import cartopy.feature as cfeature

mpl.rcParams['figure.figsize'] = 12, 8
mpl.rcParams['font.size'] = 18
```

### Load in the data using `pandas`

```python
with open('weather.csv', 'r') as f:
    data = pd.read_csv(f)
```

### Show the data, we have 15 columns, and 416937 observations

```python
data.head()
```

#### Convert dates to datetime for better plotting

```python
data['date'] = pd.to_datetime(data['date'], format='%Y%m%d')
```

### Questions to investigate:

1) 

#### First let's see where the weather stations are

```python
stationdata = data.groupby('station').mean()
```

```python
extent = [-120, -70, max(stationdata['latitude']+1), min(stationdata['latitude']-1)]
central_lon = np.mean(extent[:2])
central_lat = np.mean(extent[2:])

plt.figure(figsize=(8, 12))
ax = plt.axes(projection=ccrs.AlbersEqualArea(central_lon, central_lat))
ax.set_extent(extent)

ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.LAND, edgecolor='black')
ax.add_feature(cfeature.LAKES, edgecolor='black')
ax.add_feature(cfeature.RIVERS)
ax.gridlines()

ax.scatter(stationdata['longitude'], stationdata['latitude'], transform=ccrs.PlateCarree(),
          s=3)
```

The coverage of the US is good, although there are a disproportionately high number of stations in the northwest USA.

Now, for another sanity check, let's make sure there is a seasonal cycle to the data

```python
dailydata = data.groupby('date').mean()
count = data.groupby('date').count()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
dailydata['TAVG'].plot.line(ax=ax1)
count['TAVG'].plot.bar(ax=ax2, width=1)

ax2.set_ylabel('Count')
ax2.set_xticklabels(count.index[::33])
ax2.set_xticks(np.arange(count.shape[0], step=33))
ax2.get_xticks()
```

We do see a reasonable seasonal cycle, but notice there is only data through September 21. We also see that the number of observations collected drops off toward the end of the time period.
